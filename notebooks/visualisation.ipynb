```python
df = pd.read_csv("../data/cleaned_stocks/stock_0")
fig = go.Figure(data=[go.Candlestick(x=df['time'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
