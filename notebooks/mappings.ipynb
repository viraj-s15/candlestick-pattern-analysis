```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = [19, 15]
df = pd.read_csv("../data/no_nan/stock_0.csv")
cols = df.columns[5:]
fig, axs = plt.subplots(len(cols) // 2, len(cols) // 2)
df[cols] = df[cols].apply(np.int8)
for row in range(len(df.columns[5:]) // 2):
    for col in range(len(df.columns[5:]) // 2):
        axs[row,col].hist(df[df.columns[row + col + 5]])
        axs[row, col].set_title(df.columns[row+col+5])
        plt.tight_layout()
```
